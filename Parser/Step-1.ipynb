1. Read dataset
2. Create parser to make Dataset Dictionary for Symptoms values

```python
# Step 1: Import necessary library
import pandas as pd
import numpy as np
```

```python
# Step 2: Load the dataset
data = pd.read_csv('data.csv')
data
```

```python
# Step 3: Define a function to parse the dataset and create a dictionary of symptoms for each patient
def create_symptom_dict(data):
    # List of symptom-related columns
    symptom_columns = ['Fever_Mild', 'Fever_Low', 'Fever_High', 'Cough_Mild', 'Cough_Low', 'Cough_High',
                       'Cold_Mild', 'Cold_Low', 'Cold_High', 'Other_Symptoms']
    
    # Initialize an empty dictionary
    symptom_dict = {}
     # Step 4: Iterate over each row in the dataset
    for index, row in data.iterrows():
        patient_id = row['Patient_Id']  # Extract patient ID
        symptoms = {}  # Initialize an empty dictionary for symptoms
        
        # Step 5: Extract symptom values for each symptom-related column
        for symptom in symptom_columns:
            symptoms[symptom] = row[symptom]
        
        # Assign the dictionary of symptoms to the patient ID
        symptom_dict[patient_id] = symptoms
    
    return symptom_dict
```

```python
# Step 6: Create the symptom dictionary
symptom_dict = create_symptom_dict(data)
```

```python
# Step 7: Display the dictionary for the first 5 patients to verify
symptom_dict_sample = {key: symptom_dict[key] for key in list(symptom_dict)[:5]}
symptom_dict_sample
```
