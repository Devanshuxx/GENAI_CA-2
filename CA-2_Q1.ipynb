```python
#importing required libraries
import random
import datetime
```

```python
class Transaction:
    def __init__(self, date, amount, transaction_type):
        self.date = date
        self.amount = amount
        self.transaction_type = transaction_type
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        self.transactions = []
```

```python
def generate_accounts(num_accounts, seed_amount, num_months, max_transactions_per_month):
    accounts = []
    random.seed(42)

    for i in range(num_accounts):
        account = BankAccount(f"ACC{i+1:03d}", random.uniform(100, 10000))
        current_date = datetime.date(2024, 1, 1)

        for _ in range(num_months):
            for _ in range(random.randint(1, max_transactions_per_month)):
                amount = random.uniform(-seed_amount, seed_amount)
                account.transactions.append(Transaction(current_date, amount, "deposit" if amount > 0 else "withdrawal"))
                account.balance += amount
            current_date += datetime.timedelta(days=30)

        accounts.append(account)

    return sorted(accounts, key=lambda x: x.balance)
```

```python
# Generating and print accounts random accounts
accounts = generate_accounts(100, 1000, 6, 5)
for account in accounts:
    print(f"Account: {account.account_number}, Final Balance: ${account.balance:.2f}")
```
